```python
import cv2 #OpenCV-Python is a library of Python bindings designed to solve computer vision problems
import numpy as np #NumPy is a Python library used for working with arrays
```

```python
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("images.jpg") #cv2.imread() method loads an image from the specified file
```

```python
cv2_imshow(img)
```

```python
# Reading the black and white image
b_and_w_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Displaying Black And White Image
cv2_imshow(b_and_w_image)
cv2.waitKey(0)
```

```python
# Inverting the image
inverted_image = 255 - b_and_w_image
```

```python
# Displaying Inverted Image
cv2_imshow(inverted_image)
cv2.waitKey()
```

```python
# Blurring the image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
# Inverting the Blurred Image
inverted_blurred = 255 - blurred
```

```python
# Creating the Pencil Sketch image
pencil_sketch = cv2.divide(b_and_w_image, inverted_blurred, scale=256.0)
```

```python
# Displaying Sketch Image
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
